```python
epochs_eegnet = 30

pipelines = {}
pipelines["CSP + LDA"] = make_pipeline(
        CSP(n_components=8, reg=1e-4),  # without reg, get 'not positive definite' error 
        # CSP(n_components=2, reg=1e-4),  # without reg, get 'not positive definite' error
        LDA()
)

X, labels, meta = paradigm.get_data(dataset=ds, subjects=subject_list)
_, n_channels, n_samples = X.shape
clf = create_eegnet(n_classes, n_channels, n_samples, epochs_eegnet)
pipelines["EEGNet"] = make_pipeline(clf)
```

```python
overwrite = True
debug = True
random_state = 23
rng = np.random.default_rng(random_state)


dataname = ds.code.replace(" ", "_")
suffix = f'{dataname}_epoch{epochs_eegnet}_nclass{n_classes}'

evaluation = WithinSessionEvaluation(
        # evaluation = CrossSessionEvaluation(  # not compatible with PhysionetMI
            paradigm=paradigm, 
            datasets=[ds], 
            overwrite=overwrite,
            suffix=suffix,
            random_state=random_state,
            n_cv=8,
        )
```

```python
results = evaluation.process(pipelines)
```
